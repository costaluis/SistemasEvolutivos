```python
#include bibs
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
import numpy as np
import pandas as pd
```

```python
dados = pd.read_csv("../files/output2.csv")

ax = plt.axes(projection="3d")
ax.plot3D(dados['Fitness'], dados['Num_elementos'], dados['Iteracao'])
```

```python
y_diff_zero = dados['Fitness'] != 0

index = dados[dados['Fitness']==0].first_valid_index()

y_diff_zero[index] = True

plt.step(dados[y_diff_zero]['Iteracao'], dados[y_diff_zero]['Fitness'])

x = dados['Iteracao'][index]

x_scale = np.arange(0, x, 50)
x_scale = np.append(x_scale, x)

plt.grid(True)
_= plt.xticks(x_scale)
```
