```python
#include libs
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
import numpy as np
import pandas as pd
```

```python
dados = pd.read_csv("../files/output2.csv")
```

```python
y_diff_zero = dados['Fitness'] != 0

index = dados[dados['Fitness']==0].first_valid_index()

y_diff_zero[index] = True

plt.step(dados[y_diff_zero]['Iteracao'], dados[y_diff_zero]['Fitness'], where='post')

plt.xlabel("Iterações")
plt.ylabel("Fitness")
plt.title("Fitness x Iterações")

x = dados['Iteracao'][index]

x_scale = np.arange(0, x, 50)
x_scale = np.append(x_scale, x)

plt.grid(True)
_= plt.xticks(x_scale)
```

```python
plt.step(dados['Iteracao'], dados['Num_elementos'], where='post')

x = dados['Iteracao'].iloc[-1]

x_scale = np.arange(0, x-100000, 100000)
x_scale = np.append(x_scale, x)

plt.xlabel("Iterações")
plt.ylabel("Num. Elementos")
plt.title("Num. Elementos x Iterações")

plt.grid(True)
_= plt.xticks(x_scale)
```

```python
ax = plt.axes(projection="3d")
ax.plot3D(dados['Iteracao'], dados['Num_elementos'], dados['Fitness'])

plt.xlabel("Iterações")
plt.ylabel("Num. Elementos")
_ = ax.set_zlabel("Fitness")
```
